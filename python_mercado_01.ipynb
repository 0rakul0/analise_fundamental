o calculo do retorno Ã© dado como

ret = preco(i) - preco( i-1 ) / preco ( i-1 )

```python
x = []
i_0 = data_frame['VALOR_COTA']
for i in i_0:
    x.append(i)
x.reverse()
preco = np.array(x)
ret = (preco[1:]-preco[0:-1])/preco[0:-1]
ret
```

```python
t = np.arange(0,7)
plt.plot(t,ret)
plt.title('retorno')
plt.show()
```

```python
import math
import statistics as st

media = st.mean(ret)
desvio = st.pstdev(ret)

cen_ot = media+2*desvio/math.sqrt(7)
cen_ps = media-2*desvio/math.sqrt(7)

print(cen_ot, cen_ps)
```
